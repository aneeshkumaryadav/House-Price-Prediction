# House Price Prediction - Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel('../dataset/HousePricePredictionindore.xlsx')
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
print(df.isnull().sum())
print('Duplicates:',df.duplicated().sum())
```

```python
print(df.columns.tolist())
```

```python
num_cols=df.select_dtypes(include=['int64','float64']).columns
df[num_cols].hist(figsize=(12,8))
plt.tight_layout();plt.show()
```

```python
if 'Price' in df.columns:
 plt.figure(figsize=(6,4));plt.hist(df['Price'],bins=20);plt.title('Price Distribution');plt.show()
```

```python
for c in num_cols:
 plt.figure(figsize=(5,3));plt.boxplot(df[c],vert=False);plt.title(c);plt.show()
```

```python
corr=df[num_cols].corr();plt.figure(figsize=(8,6));plt.imshow(corr);plt.xticks(range(len(corr.columns)),corr.columns,rotation=90);plt.yticks(range(len(corr.columns)),corr.columns);plt.colorbar();plt.show()
```

```python
cat_cols=df.select_dtypes(include=['object','category','string']).columns
for c in cat_cols:
 print('\n',c);print(df[c].value_counts())
```
